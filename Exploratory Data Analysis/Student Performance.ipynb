```python
categorical_variables = ['gender','race/ethnicity','parental level of education','lunch','test preparation course']
```

```python
print(categorical_variables)
```

```python
for variables in categorical_variables:
    data[variables].fillna('Missing',inplace=True)
    dummies = pd.get_dummies(data[variables],prefix = variables)
    data = pd.concat([data,dummies],axis=1)
    data.drop([variables],axis=1,inplace=True)
```

```python
def printall(data,max_rows =10):
    from IPython.display import display,HTML
    display(HTML(data.to_html(max_rows=max_rows)))
printall(data)
```
